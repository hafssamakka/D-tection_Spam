```python
def nettoyage_de_texte(text):

    text = text.lower() # transformation en minuscule
    text = re.sub(r'<[^>]+>', '', text) # ou text = BeautifulSoup(text, "html.parser").get_text()
    text = re.sub(r'http[s]?://\S+|www\.\S+', '', text) # pour supprimer les urls
    text = emoji.demojize(text) # pour remplacer les emojis
    # Replace '@' with 'a' as requested
    text = re.sub(r'@', 'a', text)
    text = unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('utf-8') #les lettres avec accents
    # les caractères spéciaux et les chiffres, en ne gardant que les lettres et les espaces
    text = re.sub(r"[^'.a-zA-Z\s]", "", text)
    # signes de ponctuation
    text = re.sub(r'[,!?;:()*]', '', text)
    # les espaces multiples
    text = re.sub(r'\s+', ' ', text).strip()
    # text = re.sub(r'http[s]?://\S+|www\.\S+', '', text)
    text = re.sub(r"'", " ", text)
    #text = re.sub(r'\d', '', text)
    # les retours à la ligne
    text = re.sub(r'\n', '', text)
    #text = re.sub(r'<[^>]+>', '', text) # pour les balises
    #text = BeautifulSoup(text, "html.parser").get_text() # pour les balises aussi
    text = re.sub(r'\d+', '', text) #pour enlever les chiffres
    text = re.sub(r'\.{2,}', '.', text) # pour remplacer 2 points ou plus par un seul point

    return text
```

Partie 3: Stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Récupérer la liste des stopwords en français et en anglais
stop_words_french = set(stopwords.words('french'))
stop_words_english = set(stopwords.words('english'))

# Combiner les deux listes de stopwords
combined_stop_words = stop_words_french.union(stop_words_english)

# Ajouter les mots spécifiques à enlever
words_to_remove_manually = {'dghpcybpcyblbmnvzgvk','svcshredditcommenttntwyhodepthsubredditnamenlp'}
combined_stop_words.update(words_to_remove_manually)

# Mots négatifs ou importants à conserver (pour éviter la suppression en tant que stopwords)
words_to_keep_for_sentiment = {'ne', 'pas', 'ni', 'sans', 'against', 'n','because','cause','t', 'no', 'non'}
combined_stop_words = combined_stop_words.difference(words_to_keep_for_sentiment)

# Tokenisation du texte en mots
words = word_tokenize(text_norme)

# Filtrer les stopwords
filtered_words = [w for w in words if w.lower() not in combined_stop_words]

# Reconstituer le texte filtré
filtered_text = ' '.join(filtered_words)

print(filtered_text)
```

```python
import nltk
nltk.download('punkt')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

from nltk.stem import SnowballStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

# English stemmer and lemmatizer
stemmer = SnowballStemmer('english')
lemmatizer = WordNetLemmatizer()

# Tokenize the text
words = word_tokenize(filtered_text)

# Perform stemming and lemmatization on the words
stemmed_words = [stemmer.stem(word) for word in words]
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in words]

# Print the stemmed and lemmatized words
print("Stemmed words: ", stemmed_words)
```
